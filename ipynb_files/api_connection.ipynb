## Compile Roster DataFrame

```python
roster_df = pd.DataFrame()
for team in team_rosters:
    team_id = team['id']
    team_df = pd.DataFrame(team['player'])
    team_df['team_id'] = team_id
    roster_df = roster_df.append(team_df)
roster_df 
```
