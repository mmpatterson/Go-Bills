## Passing Yard Points

```python
def pass_yards(dataframe):
    for index, row in dataframe.iterrows():
        if(row['play_type'] == 'pass'):
            dataframe.at[index, 'passing_yards_points'] = row['yards_gained'] * init.pass_yd
    return dataframe
```

## Passing Completions/Incompletions

```python
def pass_comps(dataframe):
    for index, row in dataframe.iterrows():
        if(row['play_type'] == 'pass'):
            if(row['incomplete_pass'] == 1 ):
                dataframe.at[index, 'passing_comps_points'] = init.pass_incomp
            elif(row['interception'] == 1 ):
                dataframe.at[index, 'passing_comps_points'] = init.pass_incomp
            elif(row['complete_pass'] == 1):
                dataframe.at[index, 'passing_comps_points'] = init.pass_comps
        elif(row['play_type'] == 'qb_spike'):
                dataframe.at[index, 'passing_comps_points'] = init.pass_incomp
    return dataframe
```

```python
def interceptions(dataframe):
    for index, row in dataframe.iterrows():
        if(row['interception'] ==1):
            if(row['return_touchdown'] == 0):
                dataframe.at[index, 'interception_points'] = init.pass_int
            elif(row['return_touchdown'] == 1):
                dataframe.at[index, 'interception_points'] = init.pass_int + init.pass_int_td
    return dataframe
```

## Passing 2 Point Conversions

```python
def pass_2pt(dataframe):
    for index, row in dataframe.iterrows():
        if((row['two_point_attempt'] == 1) & (row['play_type'] == 'pass')):
            if(row['two_point_conv_result'] == 'success'):
                dataframe.at[index, 'pass_2pt'] = 1
    return dataframe
```

## Fumbles (Lost and Not Lost)

```python
def fumbles(dataframe):
    for index, row in dataframe.iterrows():
        if(row['fumble'] == 1):
            if(row['fumble_lost'] == 1):
                dataframe.at[index, 'fumble_points'] = init.fumble + init.fumble_lost
            elif(row['fumble_lost'] == 0):
                dataframe.at[index, 'fumble_points'] = init.fumble
    return dataframe
```

## Get Player List- This will come into play as we evaluate the plays

```python
# Valid position groups are QB, WR, RB, FB, TE
def player_names(filepath, position):
    roster_data = pd.read_csv(filepath)
    position_df = roster_data[roster_data['position'] == position].reset_index()
    position_df = position_df[['abbr_player_name', 'gsis_id']]
    return position_df
```

## Use list of player names to grab play-by-play data that involves those specific players

```python
# Only grab plays that involve players in the inputted list
# Inputs are the list from player_names and the inital dataframe from nflscrap-R
def filter_position(position_names, play_by_play_df):
    players_df = play_by_play_df[play_by_play_df['passer_player_name'].isin(position_names['abbr_player_name']) \
           | play_by_play_df['receiver_player_name'].isin(position_names['abbr_player_name']) \
           | play_by_play_df['rusher_player_name'].isin(position_names['abbr_player_name'])]
    return players_df
```

## Passing Stats for players

```python
def passing_stats(dataframe, qbs_df):
    qb_dict = {}
    for index, row in qbs_df.iterrows():
        qb = row['abbr_player_name']
        qb_id = row['gsis_id']
        print(f"Generating passing stats for {qb}.")
        dataframe_player = dataframe.loc[dataframe['passer_player_id'] == qb_id]
        dataframe_player = pass_tds(dataframe_player)
        dataframe_player = pass_yards(dataframe_player)
        dataframe_player = pass_comps(dataframe_player)
        dataframe_player = interceptions(dataframe_player)
        dataframe_player = sack_yards(dataframe_player)
        dataframe_player = pass_2pt(dataframe_player)
        dataframe_player = fumbles(dataframe_player)
        dataframe_player = pass_first_down(dataframe_player)
        qb_dict[qb] = dataframe_player
    return qb_dict
            
```

## Rushing Stats for players

```python
def rushing_stats(dataframe, rbs_df):
    rb_dict = {}
    for index, row in rbs_df.iterrows():
        rb = row['abbr_player_name']
        rb_id = row['gsis_id']
        print(f"Generating rushing stats for {rb}.")
        dataframe_player = dataframe.loc[dataframe['rusher_player_id'] == rb_id]
        dataframe_player = rush_2pt(dataframe_player)
        dataframe_player = rush_tds(dataframe_player)
        dataframe_player = rush_yards(dataframe_player)
        dataframe_player = fumbles(dataframe_player)
        dataframe_player = rush_first_down(dataframe_player)
        rb_dict[rb] = dataframe_player
    return rb_dict
```

## Receiving Stats

```python
def receiving_stats(dataframe, wrs_df):
    wr_dict = {}
    for index, row in wrs_df.iterrows():
        wr = row['abbr_player_name']
        wr_id = row['gsis_id']
        print(f"Generating receiving stats for {wr}.")
        dataframe_player = dataframe.loc[dataframe['receiver_player_id'] == wr_id]
        dataframe_player = rec_tds(dataframe_player)
        dataframe_player = rec_yards(dataframe_player)
        dataframe_player = rec_2pt(dataframe_player)
        dataframe_player = fumbles(dataframe_player)
        dataframe_player = rec_first_down(dataframe_player)
        dataframe_player = rec_rec(dataframe_player)
        wr_dict[wr] = dataframe_player
    return wr_dict
```

## Combine Passing, Rushing, and Receiving Stats

```python
def join_dfs(names_df, pass_dict, run_dict, rec_dict):
    def sort_scores(player_list):
        def takeSecond(elem):
            return elem[1]

        player_list.sort(key=takeSecond, reverse = True)
        return player_list
    
    total_dict = {}
    sort_list = []
#     print(names_df['abbr_pl'])
    for index, row in names_df.iterrows():
        player_dict = {}
        name = names_df['abbr_player_name'][index]
#         print(name)
        total_df = pass_dict[name]
        total_df = total_df.append(run_dict[name])
        total_df = total_df.append(rec_dict[name])
        if 'passing_touchdown_points' in total_df.columns:
            pass_td_points = total_df['passing_touchdown_points'].sum()
        else:
            pass_td_points = 0
        if 'passing_yards_points' in total_df.columns:
            pass_yd_points = total_df['passing_yards_points'].sum()
        else:
            pass_yd_points = 0
        if 'passing_comps_points' in total_df.columns:
            pass_comps_points = total_df['passing_comps_points'].sum()
        else:
            pass_comps_points = 0
        if 'interception_points' in total_df.columns:
            pass_int_points = total_df['interception_points'].sum()
        else:
            pass_int_points = 0
        if 'sack_points' in total_df.columns:
            pass_sack_points = total_df['sack_points'].sum()
        else:
            pass_sack_points = 0
        if 'pass_2pt' in total_df.columns:
            pass_2_pt = total_df['pass_2pt'].sum()
        else:
            pass_2_pt = 0
        if 'fumble_points' in total_df.columns:
            fumble_points = total_df['fumble_points'].sum()
        else: 
            fumble_points = 0
        if 'passing_first_down' in total_df.columns:
            pass_first_down = total_df['passing_first_down'].sum()
        else:
            pass_first_down = 0
        if 'rushing_touchdown_points' in total_df.columns:
            run_td_points = total_df['rushing_touchdown_points'].sum()
        else:
            run_td_points = 0
        if 'run_2pt' in total_df.columns:
            run_2_pt = total_df['run_2pt'].sum()
        else: 
            run_2_pt = 0
        if 'rushing_yards_points' in total_df.columns:
            run_yd_points = total_df['rushing_yards_points'].sum()
        else:
            run_yd_points = 0
        if 'rushing_first_down' in total_df.columns:
            run_first_down = total_df['rushing_first_down'].sum()
        else:
            run_first_down = 0
        if 'receiving_touchdown_points' in total_df.columns:
            rec_td_points = total_df['receiving_touchdown_points'].sum()
        else:
            rec_td_points = 0
        if 'reception_yards_points' in total_df.columns:
            rec_yd_points = total_df['reception_yards_points'].sum()
        else:
            rec_yd_points = 0
        if 'rec_2pt' in total_df.columns:
            rec_2pt = total_df['rec_2pt'].sum()
        else:
            rec_2pt = 0
        if 'receiving_first_down' in total_df.columns:
            rec_first_down = total_df['receiving_first_down'].sum()
        else:
            rec_first_down = 0
        if 'receiving_receptions' in total_df.columns:
            rec_recs = total_df['receiving_receptions'].sum()
        else:
            rec_recs = 0
        player_points = pass_td_points + pass_yd_points + pass_comps_points + pass_int_points + pass_sack_points + \
                        pass_2_pt + fumble_points + pass_first_down + run_td_points + run_2_pt + run_yd_points + \
                        run_first_down + rec_td_points + rec_yd_points + rec_2pt + rec_first_down + rec_recs
        player_dict[name] = {'name': name, 'total_points': player_points, 'dataframe': total_df, 'pass_touchdown': pass_td_points, 'pass_yards' : pass_yd_points, \
                            'passing_completions': pass_comps_points, 'interceptions': pass_int_points, \
                            'sacks' : pass_sack_points, 'pass_2_pt' : pass_2_pt, 'fumbles' : fumble_points, \
                            'pass_first_down': pass_first_down, 'run_touchdown' : run_td_points, \
                            'run_2_pt' : run_2_pt, 'run_yards' : run_yd_points, 'run_first_down': run_first_down, \
                            'rec_touchdown': rec_td_points, 'rec_yards': rec_yd_points, 'rec_2pt': rec_2pt, \
                            'rec_first_down': rec_first_down, 'rec_recs': rec_recs}
        sort_list.append((name, player_points, player_dict))
        
    sort_list = sort_scores(sort_list)
    
    for player in sort_list:
        print(player[0])
        
        total_dict[player[0]] = player[2][player[0]]
    
    return total_dict
    
```

## The Motherload

```python
def players_summary(play_by_play_df, players_list_df):
    passing_stats_df = passing_stats(play_by_play_df, players_list_df)
    rushing_stats_df = rushing_stats(play_by_play_df, players_list_df)
    receiving_stats_df = receiving_stats(play_by_play_df, players_list_df)
    final_dict = join_dfs(players_list_df, passing_stats_df, rushing_stats_df, receiving_stats_df)
    return final_dict
```

## Stack it up

```python
def stack_players(dictionary, filepath, position = "Player", rtn = 90, bar_width = 0.9, chart_width = 19, chart_height = 14):
    names = dictionary.keys()
    idx = np.asarray([i for i in range(len(names))])
    pass_td_points = []
    pass_yd_points = []
    pass_comp_points = []
    pass_int_points = []
    pass_sack_points = []
    pass_2pt_points = []
    pass_fumble_points = []
    pass_first_down_points = []
    rush_td_points = []
    rush_2pt_points = []
    rush_yards_points = []
    rush_first_down_points = []
    rec_td_points = []
    rec_yards_points = []
    rec_2pt_points = []
    rec_first_down_points = []
    rec_recs = []
    
    ind = np.arange(len(names))    # the x locations for the groups
    width = bar_width   
    labels = []
    for name in names:
        pass_td_points.append(dictionary[name]['pass_touchdown'])
        labels.append('Passing TD Points')
        pass_yd_points.append(dictionary[name]['pass_yards'])
        labels.append('Passing Yards Points')
        pass_comp_points.append(dictionary[name]['passing_completions'])
        labels.append('Passing Completion Points')
        pass_int_points.append(dictionary[name]['interceptions'])
        labels.append('Passing INT Points')
        pass_sack_points.append(dictionary[name]['sacks'])
        labels.append('Passing Sack Points')
        pass_2pt_points.append(dictionary[name]['pass_2_pt'])
        labels.append('Passing 2 Pt Points')
        pass_fumble_points.append(dictionary[name]['fumbles'])
        labels.append('Fumble Points')
        pass_first_down_points.append(dictionary[name]['pass_first_down'])
        labels.append('Passing 1st Down Points')
        rush_td_points.append(dictionary[name]['run_touchdown'])
        labels.append('Rush TD Points')
        rush_2pt_points.append(dictionary[name]['run_2_pt'])
        labels.append('Rush 2 Pt Points')
        rush_yards_points.append(dictionary[name]['run_yards'])
        labels.append('Rush Yards Points')
        rush_first_down_points.append(dictionary[name]['run_first_down'])
        labels.append('Rushing 1st Down Points')
        rec_td_points.append(dictionary[name]['rec_touchdown'])
        labels.append('Receiving TD Points')
        rec_yards_points.append(dictionary[name]['rec_yards'])
        labels.append('Receiving Yards Points')
        rec_2pt_points.append(dictionary[name]['rec_2pt'])
        labels.append('Receiving 2 Pt Points')
        rec_first_down_points.append(dictionary[name]['rec_first_down'])
        labels.append('Receiving First Down Points')
        rec_recs.append(dictionary[name]['rec_recs'])
        labels.append('Receptions')
    plt.figure(figsize=(chart_width, chart_height))
    p1 = plt.bar(ind, pass_td_points, width, color = '#e5f5f9', zorder = 1)
    pos_base = np.asarray(pass_td_points)
    p2 = plt.bar(ind, pass_yd_points, width, color = '#ccece6', bottom = pos_base)
    pos_base = pos_base + np.asarray(pass_yd_points)
    p3 = plt.bar(ind, pass_comp_points, width, color = '#99d8c9', bottom = pos_base)
    pos_base = pos_base + np.asarray(pass_comp_points)
    p4 = plt.bar(ind, pass_int_points, width, color = '#66c2a4')
    neg_base = np.asarray(pass_int_points)
    p5 = plt.bar(ind, pass_sack_points, width, color = '#41ae76', bottom = neg_base)
    neg_base = neg_base + np.asarray(pass_sack_points)
    p6 = plt.bar(ind, pass_2pt_points, width, color = '#238b45', bottom = pos_base)
    pos_base = pos_base + np.asarray(pass_2pt_points)
    p7 = plt.bar(ind, pass_fumble_points, width, color = '#006d2c', bottom = neg_base)
    neg_base = neg_base - np.asarray(pass_fumble_points)
    p8 = plt.bar(ind, pass_first_down_points, width, color = '#00441b', bottom = pos_base)
    pos_base = pos_base + np.asarray(pass_first_down_points)
    p9 = plt.bar(ind, rush_td_points, width, color = '#bdd7e7', bottom = pos_base)
    pos_base = pos_base + np.asarray(rush_td_points)
    p10 = plt.bar(ind, rush_2pt_points, width, color = '#6baed6', bottom = pos_base)
    pos_base = pos_base + np.asarray(rush_2pt_points)
    p11 = plt.bar(ind, rush_yards_points, width, color = '#3182bd', bottom = pos_base)
    pos_base = pos_base + np.asarray(rush_yards_points)
    p12 = plt.bar(ind, rush_first_down_points, width, color = '#08519c', bottom = pos_base)
    pos_base = pos_base + np.asarray(rush_first_down_points)
    p13 = plt.bar(ind, rec_td_points, width, color = '#fdbe85', bottom = pos_base)
    pos_base = pos_base + np.asarray(rec_td_points)
    p14 = plt.bar(ind, rec_yards_points, width, color = '#fd8d3c', bottom = pos_base)
    pos_base = pos_base + np.asarray(rec_yards_points)
    p15 = plt.bar(ind, rec_2pt_points, width, color = '#e6550d', bottom = pos_base)
    pos_base = pos_base + np.asarray(rec_2pt_points)
    p16 = plt.bar(ind, rec_first_down_points, width, color ='#a63603', bottom = pos_base)
    pos_base = pos_base + np.asarray(rec_first_down_points)
    p17 = plt.bar(ind, rec_recs, width, color = '#7f2704', bottom = pos_base)
    plt.title(f'2019 {position} Fantasy Breakdown, Ordered by Total Points')
    plt.xticks(idx, names, rotation = rtn)
    plt.tick_params(axis = 'both', labelsize = 'large')
    plt.figtext(0.99, 0.01, 'Plot made by @MaxFromChicago, data courtesy of nflscrap-R', horizontalalignment='right') 
    plt.tight_layout()
    plt.legend(labels)
    plt.savefig(filepath)
    plt.show()
    
```
