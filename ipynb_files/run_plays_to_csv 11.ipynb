```python
def rb_run_breakdowns(rb_plays, rb_names_df): 
    rbs_df = {}
    for rb in range(len(rb_names_df['gsis_id'])):
#         print(rb_names_df)
        rb_spec_run = rb_plays[(rb_plays['play_type'] == 'run') \
                           & (rb_plays['rusher_player_id'] == rb_names_df['gsis_id'][rb])]
        rb_spec_pass = rb_plays[(rb_plays['play_type'] == 'pass') \
                           & (rb_plays['rusher_player_id'] == rb_names_df['gsis_id'][rb])]
        
        rb_dict[rb] = 'run': {'play_id': rb_spec_run['play_id'],'play_desc': rb_spec_run['desc'], \
                              'possession_team': rb_spec_run['posteam'], 'defending_team':rb_spec_run['defteam'], \
                              'side_of_field':rb_spec_run['side_of_field'], 'yardline_100':rb_spec_run['yardline_100'], \
                    'date':rb_spec_run['game_date'], 'qtr_sec_rem':rb_spec_run['quarter_seconds_remaining'], \
                              'half': rb_spec_run['game_half'], 'drive_number': rb_spec_run['drive'], \
                              'quarter':rb_spec_run['qtr'], 'down':rb_spec_run['down'], \
                              'goal_to_go_tf':rb_spec_run['goal_to_go'], 'quarter_time':rb_spec_run['time'], \
                              'yard_line':rb_spec_run['yrdln'], 'yards_to_go':rb_spec_run['ydstogo'], \
                              'yards_net':rb_spec_run['ydsnet'], 'play_type':rb_spec_run['play_type'], \
                              'yards_gained':'yards_gained', 'shotgun_tf':rb_spec_run['shotgun'], \
                    'no_huddle_tf':rb_spec_run['no_huddle'], 'run_direction'rb_spec_run['run_location'], \
                              'run_gap':rb_spec_run['run_gap'], 'pos_team_score':rb_spec_run['posteam_score'], \
                              'def_team_score':rb_spec_run['defteam_score'], 'score_diff':rb_spec_run['score_differential'], \
                    'epa':rb_spec_run['epa'], 'wp', 'first_down_rush', 'third_down_converted', 'third_down_failed', \
                    'fourth_down_converted', 'fourth_down_failed', 'fumble', 'fumble_forced', 'fumble_not_forced', \
                    'fumble_out_of_bounds', 'solo_tackle', 'safety', 'penalty', 'tackled_for_loss', \
                    'fumble_lost', 'rush_attempt', 'lateral_rush', 'passer_player_name', 'rusher_player_id', \
                    'rusher_player_name', 'blocked_player_name', 'tackle_for_loss_1_player_name', \
                    'tackle_for_loss_2_player_name', 'forced_fumble_player_1_team', \
                    'forced_fumble_player_1_player_name', 'forced_fumble_player_1_player_id', 'forced_fumble_player_2_team', \
                    'forced_fumble_player_2_player_name', 'forced_fumble_player_2_player_id', 'solo_tackle_1_team', 
                    'solo_tackle_2_team', 'solo_tackle_1_player_id', 'solo_tackle_2_player_id', 'solo_tackle_1_player_name', 
                    'solo_tackle_2_player_name', 'fumbled_2_player_id', 'fumbled_2_player_name', 'fumble_recovery_1_team', \
                    'fumble_recovery_1_player_name', 'fumble_recovery_1_player_id', 'penalty_player_id', 'penalty_player_name', \
                    'penalty_yards', 'penalty_type'}
        rbs_df[rb_names_df['abbr_player_name'][rb]] = rb_df
    return rbs_df
```

```python
rb_dict = rb_run_breakdowns(rb_plays, rb_names_df)
```
